```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("ecommerce.db")
cursor = conn.cursor()

print("Database created successfully")
```

# Create Tables

```python
cursor.executescript("""
CREATE TABLE customers (
    customer_id INTEGER PRIMARY KEY,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    city TEXT,
    state TEXT,
    join_date DATE,
    is_premium BOOLEAN
);

CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    product_name TEXT NOT NULL,
    category TEXT,
    brand TEXT,
    unit_price REAL CHECK(unit_price > 0),
    stock_qty INTEGER
);

CREATE TABLE orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    order_date DATE,
    status TEXT,
    total_amount REAL,
    FOREIGN KEY(customer_id)
    REFERENCES customers(customer_id)
);

CREATE TABLE order_items (
    item_id INTEGER PRIMARY KEY,
    order_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    unit_price REAL,
    discount_pct REAL,
    FOREIGN KEY(order_id)
    REFERENCES orders(order_id),
    FOREIGN KEY(product_id)
    REFERENCES products(product_id)
);
""")

conn.commit()
```

## Q11. What does idx_orders_date do?

The index idx_orders_date is created on the order_date column.

Purpose:
- It speeds up searching and filtering operations on order_date.
- Instead of scanning every row in the table, SQLite can directly locate matching records through the index.
- This significantly improves performance on large datasets.

Example query that benefits from this index:

```python
pd.read_sql_query("""
SELECT *
FROM orders
WHERE order_date BETWEEN '2024-08-01' AND '2024-08-31'
""", conn)
```

Without an index, SQLite performs a full table scan.
With the index, SQLite can quickly locate matching dates.

## Q12. Would the index on join_date be used?

Query: SELECT *
FROM customers
WHERE YEAR(join_date) = 2024;

No.

Reason:
Applying a function (YEAR) to the indexed column forces the database to evaluate every row, preventing efficient index usage.

This makes the query non-SARGable.

Index-friendly version:

```python
pd.read_sql_query("""
SELECT *
FROM customers
WHERE join_date >= '2024-01-01'
AND join_date < '2025-01-01'
""", conn)
```

This version allows SQLite to use the index on join_date efficiently because the column is compared directly without applying any function.

## Q15. Average unit price of products in each category

```python
pd.read_sql_query("""
SELECT category,
       ROUND(AVG(unit_price),2) AS average_price
FROM products
GROUP BY category
""", conn)
```

AVG(unit_price) calculates the average product price for each category.

## Q16. Count of orders and total revenue by status

```python
pd.read_sql_query("""
SELECT status,
       COUNT(*) AS total_orders,
       SUM(total_amount) AS total_revenue
FROM orders
GROUP BY status
ORDER BY total_revenue DESC
""", conn)
```

This query groups orders by status, counts the orders in each group, calculates total revenue, and sorts results by revenue in descending order.

## Q17. Most expensive and cheapest product in each category

```python
pd.read_sql_query("""
SELECT category,
       MAX(unit_price) AS most_expensive,
       MIN(unit_price) AS cheapest
FROM products
GROUP BY category
""", conn)
```

MAX(unit_price) returns the highest priced product and MIN(unit_price) returns the lowest priced product in each category.

## Q18. Categories with average unit price greater than ₹2000

```python
pd.read_sql_query("""
SELECT category,
       ROUND(AVG(unit_price),2) AS average_price
FROM products
GROUP BY category
HAVING AVG(unit_price) > 2000
""", conn)
```

HAVING filters grouped results. This query returns only categories whose average unit price exceeds ₹2000.

## Q19. Display each order with customer details using INNER JOIN

```python
pd.read_sql_query("""
SELECT
    o.order_id,
    o.order_date,
    c.first_name,
    c.last_name,
    o.total_amount
FROM orders o
INNER JOIN customers c
ON o.customer_id = c.customer_id
ORDER BY o.order_id
""", conn)
```

INNER JOIN returns only matching records from both tables.
Each order is matched with the customer who placed it.


## Q20. Display all customers and their orders using LEFT JOIN

LEFT JOIN returns all customers.
If a customer has no order, order-related columns appear as NULL.

## Q21. Display product details for every order item

```python
pd.read_sql_query("""
SELECT
    o.order_id,
    p.product_name,
    oi.quantity,
    oi.unit_price,
    oi.discount_pct
FROM orders o
INNER JOIN order_items oi
ON o.order_id = oi.order_id
INNER JOIN products p
ON oi.product_id = p.product_id
ORDER BY o.order_id
""", conn)
```

This query joins three tables:
Orders → Order_Items → Products

It shows which products were purchased in each order.

## Q22. Difference between LEFT JOIN and RIGHT JOIN

LEFT JOIN:
Returns all records from the left table and matching records from the right table.

Example:
customers LEFT JOIN orders

Result:
All customers are displayed.
Customers without orders show NULL values for order columns.

RIGHT JOIN:
Returns all records from the right table and matching records from the left table.

Example:
customers RIGHT JOIN orders

Result:
All orders are displayed.
Orders without matching customers show NULL values for customer columns.

FULL OUTER JOIN:
Returns all matching and non-matching rows from both tables.

Use Case:
When you want every customer and every order, even if no matching record exists.

Note:
SQLite does not support FULL OUTER JOIN directly.

## Q24. Classify products into price tiers using CASE

```python
pd.read_sql_query("""
SELECT
    product_name,
    unit_price,
    CASE
        WHEN unit_price < 1000 THEN 'Budget'
        WHEN unit_price BETWEEN 1000 AND 3000 THEN 'Mid-Range'
        ELSE 'Premium'
    END AS price_tier
FROM products
ORDER BY unit_price
""", conn)
```

Products are classified into:

- Budget → unit_price < 1000
- Mid-Range → unit_price BETWEEN 1000 AND 3000
- Premium → unit_price > 3000

## Q25. Count Delivered vs Not Delivered Orders

```python
pd.read_sql_query("""
SELECT
    SUM(
        CASE
            WHEN status = 'Delivered' THEN 1
            ELSE 0
        END
    ) AS delivered_orders,

    SUM(
        CASE
            WHEN status <> 'Delivered' THEN 1
            ELSE 0
        END
    ) AS not_delivered_orders

FROM orders
""", conn)
```

CASE statements are used inside aggregate functions.

Delivered orders are counted separately from all other statuses
(Shipped, Pending, Cancelled).

## Q26. ACID Properties

### A - Atomicity
A transaction is treated as a single unit.
Either all operations succeed or all fail.

Example:
If ₹1000 is transferred from Account A to Account B,
both debit and credit must happen together.
If one fails, everything is rolled back.

---

### C - Consistency
A transaction must take the database from one valid state to another valid state.

Example:
If total money in the banking system is ₹10,000 before transfer,
it should remain ₹10,000 after transfer.

---

### I - Isolation
Multiple transactions should not interfere with each other.

Example:
Two users transferring money simultaneously should not see
incomplete changes made by the other transaction.

---

### D - Durability
Once a transaction is committed, the changes are permanently saved.

Example:
After a successful bank transfer, the updated balances remain
even if the database server crashes immediately afterward.

---

### Real-World Example: Bank Transfer

1. Debit ₹1000 from Account A.
2. Credit ₹1000 to Account B.
3. Commit transaction.

Atomicity:
Both operations happen together.

Consistency:
Total money remains unchanged.

Isolation:
Other users cannot see partial updates.

Durability:
After commit, the transfer remains permanent.

## Q27. Transaction Example Using BEGIN, COMMIT and ROLLBACK

```python
transaction_sql = """
BEGIN TRANSACTION;

INSERT INTO orders
VALUES
(
    1011,
    102,
    '2024-09-01',
    'Pending',
    1598.00
);

INSERT INTO order_items
VALUES
(
    5016,
    1011,
    202,
    1,
    799.00,
    0
);

INSERT INTO order_items
VALUES
(
    5017,
    1011,
    208,
    1,
    799.00,
    0
);

UPDATE products
SET stock_qty = stock_qty - 1
WHERE product_id = 202;

UPDATE products
SET stock_qty = stock_qty - 1
WHERE product_id = 208;

COMMIT;
"""
```

```python
try:
    conn.execute("BEGIN TRANSACTION")

    conn.execute("""
    INSERT INTO orders
    VALUES
    (1011,102,'2024-09-01','Pending',1598.00)
    """)

    conn.execute("""
    INSERT INTO order_items
    VALUES
    (5016,1011,202,1,799.00,0)
    """)

    conn.execute("""
    INSERT INTO order_items
    VALUES
    (5017,1011,208,1,799.00,0)
    """)

    conn.execute("""
    UPDATE products
    SET stock_qty = stock_qty - 1
    WHERE product_id = 202
    """)

    conn.execute("""
    UPDATE products
    SET stock_qty = stock_qty - 1
    WHERE product_id = 208
    """)

    conn.commit()

    print("Transaction committed successfully")

except Exception as e:
    conn.rollback()
    print("Transaction rolled back")
    print(e)
```

Transaction Flow:

1. Insert a new order.
2. Insert two order items.
3. Reduce stock quantity.
4. If every step succeeds → COMMIT.
5. If any step fails → ROLLBACK.

This ensures Atomicity and Consistency.
