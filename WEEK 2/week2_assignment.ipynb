###  Create SQLite Database

```python
conn = sqlite3.connect("superstore.db")
```

```python
df.to_sql(
    "superstore",
    conn,
    if_exists="replace",
    index=False
)
```

### Verify SQL Table

```python
pd.read_sql("""
SELECT *
FROM superstore
LIMIT 5
""", conn)
```

# SQL Analysis

## Task 3: Filtering Data Using WHERE Clause

### Filter Records from West Region

```python
pd.read_sql("""
SELECT *
FROM superstore
WHERE Region='West'
LIMIT 10
""", conn)
```

### Filter Furniture Category

```python
pd.read_sql("""
SELECT *
FROM superstore
WHERE Category='Furniture'
LIMIT 10
""", conn)
```

### Transactions with Sales Greater Than 1000

```python
pd.read_sql("""
SELECT *
FROM superstore
WHERE Sales > 1000
""", conn)
```

### Filter Orders from Year 2016

```python
pd.read_sql("""
SELECT *
FROM superstore
WHERE [Ship Date] LIKE '%2016'
LIMIT 10
""", conn)
```

## Task 4: Aggregations Using GROUP BY

### Total Sales by Category

```python
pd.read_sql("""
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Category
""", conn)
```

### Average Sales by Category

```python
pd.read_sql("""
SELECT Category,
       AVG(Sales) AS Avg_Sales
FROM superstore
GROUP BY Category
""", conn)
```

### Total Quantity Sold

```python
pd.read_sql("""
SELECT Category,
       SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY Category
""", conn)
```

## Task 5: Sorting and Limiting Results

### Top 10 Products

```python
pd.read_sql("""
SELECT [Product Name],
       SUM(Sales) AS TotalSales
FROM superstore
GROUP BY [Product Name]
ORDER BY TotalSales DESC
LIMIT 10
""", conn)
```

### Top Categories

```python
pd.read_sql("""
SELECT Category,
       SUM(Sales) AS TotalSales
FROM superstore
GROUP BY Category
ORDER BY TotalSales DESC
""", conn)
```

## Task 6: Business Use Cases

### Use Case 1: Monthly Sales Trends

```python
pd.read_sql("""
SELECT [Order Date]
FROM superstore
LIMIT 10
""", conn)
```

### Top 10 Customers by Sales

```python
pd.read_sql("""
SELECT [Customer Name],
       SUM(Sales) AS TotalSales
FROM superstore
GROUP BY [Customer Name]
ORDER BY TotalSales DESC
LIMIT 10
""", conn)
```

### Detect Duplicate Records

```python
pd.read_sql("""
SELECT [Row ID],
       COUNT(*) AS Duplicate_Count
FROM superstore
GROUP BY [Row ID]
HAVING COUNT(*) > 1
""", conn)
```

## Task 5: Data Validation

### Row Count

```python
pd.read_sql("""
SELECT COUNT(*) AS Total_Rows
FROM superstore
""", conn)
```

### Sales Data Quality

```python
pd.read_sql("""
SELECT MIN(Sales) AS MinSales,
       MAX(Sales) AS MaxSales,
       AVG(Sales) AS AvgSales
FROM superstore
""", conn)
```

### Profit Data Quality

```python
pd.read_sql("""
SELECT MIN(Profit) AS MinProfit,
       MAX(Profit) AS MaxProfit
FROM superstore
""", conn)
```
