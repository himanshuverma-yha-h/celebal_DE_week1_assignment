TASK -3 HANDLE MISSING VALUES

identifies the missing values

```python
df.isnull().sum()
```

fill the missing values

```python
df["discount"] = df["discount"].fillna(df["discount"].mean())
```

verfying if the missing values are filled with mean or not

```python
df["discount"].isnull().sum()
```

for text columns fill the missing values with not available

```python
df["what_customers_said"] = df["what_customers_said"].fillna("Not Available")
df["seller_name"] = df["seller_name"].fillna("Not Available")
df["videos"] = df["videos"].fillna("Not Available")
df["seller_information"] = df["seller_information"].fillna("Not Available")
df["variations"] = df["variations"].fillna("Not Available")
```

hence verified all the missing values are filled and now we dont have any missing value

```python
df.isnull().sum()
```

we could also drop the missing values instead of filling them 

```python
df.dropna()
```

TASK -5 REMOVE DUPLICATES

TASK -6 CREATING A DREIVED COLUMN

checking the data type for making the derived column

```python
df[["initial_price", "discount"]].head()
```

now we will make a column which will give us a discount amout which is derived from initial price and discount

```python
df["discount_amount"] = df["initial_price"] * df["discount"] / 100
```

verifying the derived column

```python
df[["initial_price", "discount", "discount_amount"]].head()
```

TASK -7 SAVEING THE CLEAN DATASET
